```python
H_dist={
    'A':5,'B':3,'C':4,'D':5,'G':5,'H':7,'E':4,'F':4,'G':5,'I':0
}
H_dist1={
    'A':5,'B':3,'C':4,'D':5,'G':5,'H':7,'E':4,'F':4,'G':5,'I':0
}
NH={}
NH=H_dist
Visited=[]
label={}
parent={}
label={'A':'NS','B':'NS','C':'NS','D':'NS','E':'NS','F':'NS','G':'NS','H':'NS','I':'NS'}
'''H_dist1={
    'A':5,'B':3,'C':4,'D':5,'G':5,'H':7,'E':4,'F':4,'G':5,'I':0
}'''
l='NS'



def AO_star(node):
    global l
    val=0
    print('************************')
    print('Expanding',node)
    Visited.append(node)
    if Graph_d[node]==None:
        NH[node]=H_val(node)+1
        print(node,'return',H_val(node)+1)
        if H_dist1[node]==0:
            l='S'
        return [H_val(node)+1,l]
    else:
        for SUCC in Graph_d[node]:
            print('SUCC',SUCC)
            if len(SUCC) ==1:
                v,l=AO_star(SUCC)
                val=val+v
                label[SUCC]=l                
                NH[node]=val
                parent[SUCC]=node
                print('H:',H_dist[SUCC])
                print(node,'returns',val)                
            else:
                i=0
                for n in SUCC:
                    v,l=AO_star(n)
                    val=val+v
                    label[n]=l
                    parent[n]=node                    
                    print(SUCC,':',i)
                NH[node]=val
                print(node,'returns',val)
        NH[node]=val
        return val,l
```
